.toc - access outlines, bookmarksm are a structured list of links within a PDF document that allow users to quickly navigate to specific sections. Extracts these bookmarks as list of lists
The structure is: [lvl, title, page, ...]

lvl: The hierarchy level (e.g., 1 for top-level, 2 for sub-section, etc.)

title: The text/title of the outline entry.

page: The page number the entry points to (note: this is 1-based, so page 1 is the first page).
[
  [1, "Chapter 1", 1],
  [2, "Section 1.1", 2],
  [1, "Chapter 2", 10]
]

```python
import fitz  # PyMuPDF
import re
from pydantic import BaseModel, ValidationError, EmailStr
from typing import Literal,Optional
import logging
logging.basicConfig(
    filename="plog.log",
    level=logging.INFO,
    format='%(asctime)s %(levelname)s:%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
logging.info("Starting PDF Extraction Pipeline")# Use logging.info() or logging.error() as needed
logging.info("Starting PDF Extraction Pipeline")
# 1. Define your Pydantic model
class Lowen(BaseModel):
    name: str
    email: EmailStr
    phone: Optional[str]
    type_of_graphic: Literal["Vehicle Wrap", "Decal", "Banner"]
    dimensions: str
    quantity: int
    color_scheme: str
    additional_notes: Optional[str]

# 2. Extract text from PDF
def extract_text_from_pdf(pdf_path: str) -> str:
    doc = fitz.open(pdf_path)
    full_text = ""
    for page in doc:
        full_text += page.get_text()
    logging.info("PDF reading completed")
    return full_text

# 3. Parse fields from text using regex
def parse_fields(text: str) -> dict:
    patterns = {
    "name": r"Name:\s*(.*?)(?=\s*(Email:|Phone:|Type of Graphic:|Dimensions:|Quantity:|Color Scheme:|Additional Notes:|$))",
    "email": r"Email:\s*([\w\.-]+@[\w\.-]+)(?=\s*(Name:|Phone:|Type of Graphic:|Dimensions:|Quantity:|Color Scheme:|Additional Notes:|$))",
    "phone": r"Phone:\s*(.*?)(?=\s*(Name:|Email:|Type of Graphic:|Dimensions:|Quantity:|Color Scheme:|Additional Notes:|$|\n))"
    ,"type_of_graphic": r"Type of Graphic:\s*(.*?)(?=\s*(Name:|Email:|Phone:|Dimensions:|Quantity:|Color Scheme:|Additional Notes:|$))",
    "dimensions": r"Dimensions:\s*(.*?)(?=\s*(Name:|Email:|Phone:|Type of Graphic:|Quantity:|Color Scheme:|Additional Notes:|$))",
    "quantity": r"Quantity:\s*(\d+)(?=\s*(Name:|Email:|Phone:|Type of Graphic:|Dimensions:|Color Scheme:|Additional Notes:|$))",
    "color_scheme": r"Color Scheme:\s*(.*?)(?=\s*(Name:|Email:|Phone:|Type of Graphic:|Dimensions:|Quantity:|Additional Notes:|$))",
    "additional_notes": r"Additional Notes:\s*(.*?)(?=\s*(Name:|Email:|Phone:|Type of Graphic:|Dimensions:|Quantity:|Color Scheme:|$|\n))"


}

    extracted_data = {}
    for field, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE)
        extracted_data[field] = match.group(1).strip() if match else None
    return extracted_data

# 4. Clean and convert extracted data
def clean_data(data: dict) -> dict:
    # Normalize phone: remove non-digit characters, but keep as string
    if data.get("phone"):
        phone_digits = re.sub(r"\D", "", data["phone"])
        data["phone"] = phone_digits if phone_digits else None
    # Convert quantity to int if possible
    if data.get("quantity"):
        try:
            data["quantity"] = int(data["quantity"])
        except ValueError:
            data["quantity"] = None
    return data


# 5. Validate data with Pydantic
def validate_data(data: dict):
    try:
        validated = Lowen(**data)
        return validated.model_dump(), None
    except ValidationError as e:
        errors = {error['loc'][0]: error['msg'] for error in e.errors()}
        return None, errors

# 6. Pipeline function
def pdf_to_validated_json(pdf_path: str):
    text = extract_text_from_pdf(pdf_path)
    extracted = parse_fields(text)
    cleaned = clean_data(extracted)
    validated, errors = validate_data(cleaned)
    if validated:
        return validated
    else:
        logging.info(f"validation_errors: {errors}")

        return {"validation_errors": errors}

# Usage example
pdf_path = "/home/kronos/Desktop/resume/dummy_fleet_graphics_quote.pdf"
result = pdf_to_validated_json(pdf_path)
print(result)
logging.info("Completed the execution of the pipeline")
```
